```python
import schedule
import time
import pandas as pd
import random
from faker import Faker
from datetime import datetime
```

```python
fake = Faker('pt_BR')

Quantidades_de_linhas = 100
```

```python
def obter_produto_e_preco():
    produtos = {
        'Notebook': 3500,
        'Smartphone': 2000,
        'Monitor': 700,
        'Teclado': 300,
        'Mouse': 150
    }
    nome_produto = random.choice(list(produtos.keys()))
    preco_produto = produtos[nome_produto]
    return nome_produto, preco_produto
```

```python
def escolher_filial(): 
    locais = {
        'José': 'Barueri',
        'Amanda': 'Barueri',
        'Augusto': 'Carapicuiba',
        'Fernanda': 'Carapicuiba',
        'Antonio': 'Osasco',
        'Izabely': 'Osasco',
        'Gustavo': 'São Paulo',
        'Ana': 'São Paulo'
    }
    vendedor = random.choice(list(locais.keys()))
    cidade = locais[vendedor]
    return vendedor, cidade
```

```python

def gerar_relatorio_caotico():
    print(f"Iniciando geração de dados sujos: {datetime.now()}")
    
    dados_lista = [] 

    for _ in range(Quantidades_de_linhas):
        nome_produto, preco_produto = obter_produto_e_preco()
        vendedor, cidade = escolher_filial()
        
        if random.random() < 0.05:
            preco_produto = preco_produto * -1 

        nome_cliente = fake.name()
        if random.random() < 0.10:
            nome_cliente = None 

        if random.random() < 0.10:
            cidade = cidade.upper() 
        elif random.random() < 0.10:
            cidade = " " + cidade + " " 

        data_nasc = fake.date_of_birth(minimum_age=18, maximum_age=80)
        if random.random() < 0.20:
            data_formatada = data_nasc.strftime('%Y-%m-%d') 
        else:
            data_formatada = data_nasc.strftime('%d/%m/%Y') 

        perfil = {
            'Nome': nome_cliente,
            'CPF': fake.cpf(),
            'Data de nascimento': data_formatada,
            'Telefone': fake.cellphone_number(),
            'Cidade': cidade,
            'Produto': nome_produto,
            'Valor Gasto': preco_produto,
            'Vendedor': vendedor
        }
        dados_lista.append(perfil)
    
    linhas_duplicadas = random.choices(dados_lista, k=10)
    dados_lista.extend(linhas_duplicadas)
    df = pd.DataFrame(dados_lista)
    
    nome_arquivo = f"relatorio_sujo_{datetime.now().strftime('%d-%m-%Y_%H-%M-%S')}.xlsx"
    
    df.to_excel(nome_arquivo, index=False)
    
    print(f"Arquivo COM ERROS '{nome_arquivo}' salvo com sucesso! (Total linhas: {len(df)})")
```

```python

schedule.every(5).seconds.do(gerar_relatorio_caotico)

print("Robô sabotador iniciado...")

while True:
    schedule.run_pending()
    time.sleep(1)
```
