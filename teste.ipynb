```python
import schedule
import time
import pandas as pd
import random
from faker import Faker
from datetime import datetime
```

```python
fake = Faker('pt_BR')

Quantiades_de_linhas = 100

# --- FUNÇÕES AUXILIARES ---
def obter_produto_e_preco():
    produtos = {
        'Notebook': 3500,
        'Smartphone': 2000,
        'Monitor': 700,
        'Teclado': 300,
        'Mouse': 150
    }
    nome_produto = random.choice(list(produtos.keys())) # Escolhe um produto aleatoriamente
    preco_produto = produtos[nome_produto] # Obtém o preço do produto escolhido
    return nome_produto, preco_produto
```

```python

def escolher_filial(): 
    locais = {
        'José': 'Barueri',
        'Amanda': 'Barueri',
        'Augusto': 'Carapicuiba',
        'Fernanda': 'Carapicuiba',
        'Antonio': 'Osasco',
        'Izabely': 'Osasco',
        'Gustavo': 'São Paulo',
        'Ana': 'São Paulo'
    }

    vendedor = random.choice(list(locais.keys()))# Escolhe um vendedor aleatoriamente
    cidade = locais[vendedor] # Obtém a cidade do vendedor escolhido
    return vendedor, cidade
```

```python
def gerar_relatorio_completo():

    dados_lista = []

    # Loop para gerar as 100 linhas
    for _ in range(Quantiades_de_linhas):
        nome_produto, preco_produto = obter_produto_e_preco()
        vendedor, cidade = escolher_filial()

        perfil = {
            'Nome': fake.name(),
            'CPF': fake.cpf(),
            'Data de nascimento': fake.date_of_birth(minimum_age=18, maximum_age=80).strftime('%d/%m/%Y'),
            'Telefone': fake.cellphone_number(),
            'Cidade': cidade,
            'Produto': nome_produto,
            'Valor Gasto': preco_produto,
            'Vendedor': vendedor
        }
        dados_lista.append(perfil)
    
    df = pd.DataFrame(dados_lista)
    nome_arquivo = f"relatorio_{datetime.now().strftime('%d-%m-%Y_%H-%M')}.xlsx"
    df.to_excel(nome_arquivo, index=False,)
    
    print(f"Arquivo salvo: {nome_arquivo}")
```

```python
if __name__ == "__main__":
    gerar_relatorio_completo()
```

```python
schedule.every(20).seconds.do(gerar_relatorio_completo)

while True:
    schedule.run_pending()
    time.sleep(1)
```
